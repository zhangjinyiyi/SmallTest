```julia
using DifferentialEquations
using Plots
using Sundials
```

```julia
using BenchmarkTools
```

```julia
function lorenz!(du,u,p,t)
 du[1] = 10.0*(u[2]-u[1])
 du[2] = u[1]*(28.0-u[3]) - u[2]
 du[3] = u[1]*u[2] - (8/3)*u[3]
end

u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
prob = ODEProblem(lorenz!,u0,tspan)
```

```julia
@btime sol = solve(prob, reltol=1e-8, abstol=1e-8)
```

```julia
@btime sol = solve(prob, CVODE_BDF(), reltol=1e-8, abstol=1e-8)
```
